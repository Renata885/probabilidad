```python
import matplotlib.pyplot as plt
import math
import numpy as np

# Valores de la variable aleatoria (premios en pesos)
x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

# Verificar que las probabilidades sumen 1
print("Suma de probabilidades:", sum(p))

# Graficar distribución de probabilidad
plt.stem(x, p)
plt.title("Distribución de probabilidad del monto de premios")
plt.xlabel("Monto del premio (pesos)")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
# Calcular valor esperado E(X)
valor_esperado = 0
for i in range(len(x)):
    xi = x[i]
    pi = p[i]
    producto = xi * pi
    valor_esperado += producto
    print(f"x = {xi}, p = {pi}, x*p = {producto:.2f}")

print(f"\nValor esperado (Esperanza) = {valor_esperado:.2f} pesos")
```

```python
# Calcular E(X²)
esperanza_cuadrado = 0
for i in range(len(x)):
    esperanza_cuadrado += x[i]**2 * p[i]

# Calcular varianza y desviación estándar
varianza = esperanza_cuadrado - valor_esperado**2
desviacion = math.sqrt(varianza)

# Mostrar resultados
print(f"E(X) = {valor_esperado:.2f}")
print(f"E(X²) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación estándar = {desviacion:.2f}")
```

```python
# Distribución acumulada F(x)
F = np.cumsum(p)
print("\nDistribución acumulada F(x):")
for xi, fi in zip(x, F):
    print(f"x = {xi}, F(x) = {fi:.2f}")

# Graficar función de distribución acumulada
plt.step(x, F, where="post", color="green")
plt.title("Función de distribución acumulada F(x)")
plt.xlabel("Monto del premio (pesos)")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.show()
```

• ¿Qué significa el valor esperado en este contexto?

El valor esperado indica que en promedio el jugador gana 8.5 pesos por boleto después de muchas jugadas.

• ¿El juego es favorable para el jugador?
Si el boleto cuesta menos de 8.5 pesos, el juego es favorable.

Si cuesta más de 8.5 pesos, el juego no es favorable (el jugador pierde dinero en promedio).
