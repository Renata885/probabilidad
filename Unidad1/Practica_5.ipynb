```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("autos2 .xlsx") # Changed to read_excel and correct filename
print(df.head())
```

```python
print(df['price'].describe())
print("Valor mínimo de precios:", df['price'].min())
print("Valor máximo de precios:", df['price'].max())
print("Moda de precios:", df['price'].mode()[0])
print("Media de precios:", df['price'].mean())
print("Mediana de precios:", df['price'].median())
print("Desviación estándar de precios:", df['price'].std())
n = len(df['price'])
bins = int(1 + 3.322 * np.log10(n))
print("Cantidad óptima de bins (Sturges):", bins)
```

### CONCLUSIÓN:Estadísticas descriptivas básicas y medidas de tendencia central

El valor mínimo de los precios corresponde a los autos más económicos,
mientras que el máximo muestra la presencia de vehículos de lujo o gama alta.
La media resulta mayor que la mediana, lo que indica que la distribución
está sesgada hacia la derecha (hay autos con precios muy altos que elevan la media).
La moda representa el precio más frecuente en el conjunto de datos,
y la desviación estándar muestra que los precios tienen una gran dispersión.

```python
plt.figure(figsize=(10,6))
plt.hist(df['price'], bins=bins, color="skyblue", edgecolor="black")
plt.title("Histograma de Precios de Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### CONCLUSIÓN:Histograma de precios de los automóviles

El histograma refleja que la mayoría de los vehículos se concentran en rangos de precios bajos y medios,mientras que hay pocos autos en los rangos más altos.
Esto confirma que existen valores atípicos (outliers) de autos muy caros en el dataset.


```python
plt.figure(figsize=(10,6))
sns.boxplot(x="body-style", y="price", data=df, hue="body-style", palette="Set2", legend=False)
plt.title("Diagrama de Caja: Precio vs Tipo de Vehículo (Body Style)")
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()
```

### CONCLUSIÓN:body style
El diagrama de caja muestra que los sedanes suelen tener precios más bajos y con poca variabilidad,
mientras que los convertibles y hatchback presentan algunos valores atípicos de precios altos.
En general, el tipo de carrocería sí influye en el rango de precios.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="drive-wheels", y="price", data=df, hue="drive-wheels", palette="Set2", legend=False) # Changed x to 'drive-wheels'
plt.title("Diagrama de Caja: Precio vs Tipo de Tracción (Drive Wheels)")
plt.xlabel("Tipo de Tracción")
plt.ylabel("Precio")
plt.show()
```

### CONCLUSIÓN:drive wheels


Los vehículos con tracción trasera (rwd) presentan precios más altos en promedio,mientras que los de tracción delantera (fwd) son más económicos.
Esto puede estar relacionado con que los autos deportivos y de lujo suelen ser rwd.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="engine-location", y="price", data=df, hue="engine-location", palette="Set2", legend=False) # Changed x to 'engine-location'
plt.title("Diagrama de Caja: Precio vs Ubicación del Motor (Engine Location)")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()
```

### CONCLUSIÓN:engine-location

La mayoría de los autos tienen el motor en la parte delantera y abarcan un rango amplio de precios,
mientras que los pocos autos con motor trasero presentan precios mucho más altos.
Esto indica que la ubicación del motor es un factor asociado a vehículos exclusivos o de gama alta.
