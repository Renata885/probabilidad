```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import seaborn as sns

sns.set(style="whitegrid")


# Para cada grupo

def analizar_grupo(archivo, nombre, respuestas):
    print(f"              RESULTADOS {nombre}")
    print(f"___________________________________________\n")

    # 1 Cargar datos
    df = pd.read_csv(archivo)
    print("\nFilas del archivo excel:")
    print(df.head())
    print("\nCantidad total de filas cargadas:", len(df))

    #  2 descripcion estad
    print("\nDescripción estadística:")
    print(df.describe())

    # 3. histograma
    plt.figure(figsize=(6,4))
    plt.hist(df["Calificacion"], bins=10, color="skyblue", edgecolor="black")
    plt.title(f"Histograma de Calificaciones – {nombre}")
    plt.xlabel("Calificación Final")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    # Regresión lineal
    X = df[["HorasEstudio"]]
    y = df["Calificacion"]

    modelo = LinearRegression()
    modelo.fit(X, y)

    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_

    print(f"\nEcuación de la línea de regresión:")
    print(f"Calificación = {pendiente:.4f} * Horas + {intercepto:.4f}")

    # Correlación
    r, p_value = pearsonr(df["HorasEstudio"], df["Calificacion"])
    print(f"\nCoeficiente de Correlación de Pearson: {r:.4f}")
    print(f"Valor-p: {p_value:.6f}")

    # Gráfica
    plt.figure(figsize=(7,5))
    plt.scatter(df["HorasEstudio"], df["Calificacion"], color="blue", s=70, label="Datos")

    # línea de regresión
    x_linea = np.linspace(df["HorasEstudio"].min(), df["HorasEstudio"].max(), 100).reshape(-1,1)
    y_linea = modelo.predict(x_linea)
    plt.plot(x_linea, y_linea, color="red", linewidth=2, label="Línea de Regresión")

    plt.title(f"Regresión Lineal – {nombre}")
    plt.xlabel("Horas de Estudio")
    plt.ylabel("Calificación Final")
    plt.legend()
    plt.grid(True)
    plt.show()

 # respuestas a las preguntas

    print("\nRESPUESTAS A LAS PREGUNTAS ")
    print(f"___________________________________________\n")
    print(respuestas)


#resultados en tres grupos

respuestas_grupo1 = """
1) Interpretación del coeficiente de regresión:
En el Grupo 1, la pendiente fue negativa (aprox. –0.10), lo que indica que aumentar las horas de estudio
no mejora la calificación; incluso baja ligeramente, aunque ese efecto es mínimo.

2) Interpretación del coeficiente de correlación:
El coeficiente de correlación es casi cero (r ≈ –0.05), lo que significa que prácticamente no existe
una relación entre el tiempo de estudio y las calificaciones. La relación es muy débil.

3) ¿El modelo es estadísticamente significativo?
No, porque el valor p fue mayor a 0.05. La relación no es significativa en este grupo.

4) Confiabilidad del modelo:
El modelo no es confiable para predecir calificaciones en este grupo, ya que no hay relación clara
entre estudiar y obtener una mejor calificación.
"""

respuestas_grupo2 = """
1) Interpretación del coeficiente de regresión:
En el Grupo 2, la pendiente fue positiva y cerca de 4.22. Esto significa que por cada hora adicional
de estudio, la calificación aumenta alrededor de 4 puntos en promedio.

2) Interpretación del coeficiente de correlación:
El coeficiente de correlación fue r ≈ 0.86, lo cual muestra una relación fuerte y positiva. Aquí sí se nota
claramente que estudiar más mejora las calificaciones.

3) ¿El modelo es estadísticamente significativo?
Sí, el valor p fue muy pequeño (menor a 0.05), lo que indica que la relación es realmente significativa.

4) Confiabilidad del modelo:
El modelo es bastante confiable para hacer predicciones en este grupo, ya que la relación es fuerte
y consistente.
"""

respuestas_grupo3 = """
1) Interpretación del coeficiente de regresión:
En el Grupo 3, la pendiente fue de aproximadamente 5.28, lo que indica que cada hora extra de estudio
eleva la calificación alrededor de 5 puntos, mostrando un efecto más fuerte que en los otros grupos.

2) Interpretación del coeficiente de correlación:
El valor r ≈ 0.94 muestra una relación muy fuerte y positiva. Aquí casi todos siguen la tendencia:
más estudio, mejor calificación.

3) ¿El modelo es estadísticamente significativo?
Sí, el valor p fue extremadamente bajo, confirmando que la relación es totalmente significativa.

4) Confiabilidad del modelo:
Es el grupo con mejor confiabilidad. La relación es muy fuerte y el modelo hace buenas predicciones
basadas en las horas de estudio.
"""

#ejecucion

analizar_grupo("grupo1.csv", "GRUPO 1", respuestas_grupo1)
analizar_grupo("grupo2.csv", "GRUPO 2", respuestas_grupo2)
analizar_grupo("grupo3.csv", "GRUPO 3", respuestas_grupo3)

# conclusion general
print("            CONCLUSIONES GENERALES")
print(f"___________________________________________\n")

print("""
Después de analizar los tres grupos, se observa que la relación entre el tiempo de estudio
y las calificaciones no es igual para todos. En el Grupo 1 prácticamente no existe relación,
lo que sugiere que otros factores influyen más en el rendimiento.

En cambio, en los Grupos 2 y 3 sí se encontró una relación fuerte y positiva, donde estudiar
más se refleja en calificaciones más altas. Además, los modelos en estos grupos resultan
estadísticamente significativos y permiten hacer predicciones confiables.

En resumen, estudiar más generalmente ayuda a mejorar las calificaciones, pero este efecto
puede variar bastante dependiendo del grupo de estudiantes y de la consistencia de sus datos.
""")
```
