```python
import matplotlib.pyplot as plt
#valores de la variable aletoria
x=[0,1,2,3,4,5]
p=[0.10,0.25,0.30,0.20,0.10,0.05]
#verificar suma=1
print("suma de probabilidades:",sum(p))
#grafucar distribucion
plt.stem(x,p)
plt.title("Distribucion de probabilidad")
plt.xlabel("Num. de celulas anormales")
plt.ylabel("probabilidad")
plt.grid(True)
plt.show()
```

```python
#Iniciar una variable para acumular la suma
valor_esperado=0

#Recorre sus valores y sus probabilidades
for i in range(len(x)):
    xi=x[i]#valor de la variable aleatoria
    pi=p[i]#probabilidad de ese valor
    producto = xi * pi #multiplicar valor por su probabilidad
    valor_esperado += producto #sumar al acumulador
    #mostrar cada paso
    print (f"x = {xi}, p= {pi}, x*p = {producto:.2f}")
#mostrar resultado final
print (f"\nValor esperado (Esperanza): {valor_esperado: .2f}")
```

```python
import math
#calcular E(X)
esperanza=0
for i in range(len(x)):
  esperanza+=x[i]*p[i]
#calcular E(x2)
esperanza_cuadrado=0
for i in range(len(x)):
  esperanza_cuadrado+=x[i]**2*p[i]


#calcular varianza y desviacion estandar
varianza=esperanza_cuadrado-esperanza**2
desviacion=math.sqrt(varianza)


#Mostrar resultados
print(f"E(X)={esperanza:.2f}")
print(f"E(X2)={esperanza_cuadrado:.2f}")
print(f"Varianza={varianza:.2f}")
print(f"Desviacion estandar={desviacion:.2f}")
```

```python
#distrrubucion acumulada
import math
import numpy as np
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
p=[0.10,0.25,0.30,0.20,0.10,0.05]
F=np.cumsum(p)
for xi,fi in zip(x,F):
  print(f"x={xi},F={fi:.2f}")
#graficar
plt.step(x,F,where="post")
plt.title("funcion de distribucion acumulada")
plt.xlabel("Num. de celulas anormales")
plt.ylabel("probabilidad")
plt.grid(True)
plt.show()
```
